```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Reading the file
data = pd.read_csv("G:/Data Science ML AI\Python files/Projects/Visualisation For Company Stakeholders/Loan_prediction.csv")
data
```

```python
#Creating a new variable to store the value counts
loan_status = data['Loan_Status'].value_counts()
print(loan_status)
```

```python
#Plotting bar plot for Loan Status
loan_status.plot(kind = 'bar')
```

```python
#Plotting an unstacked bar plot for Property_Area and Loan_Status 
property_and_loan = data.groupby(['Property_Area','Loan_Status']).size().unstack()
property_and_loan
```

```python
#Plotting an unstacked bar plot for Property_Area and Loan_Status 
property_and_loan.plot(kind = 'bar', stacked = False)

#Changing the x-axis label
plt.xlabel('Property Area')

#Changing the y-axis label
plt.ylabel('Loan Status')

#Rotating the ticks of X-axis
plt.xticks(rotation = 45)
```

```python
#Plotting a stacked bar plot to compare education status and Loan status
education_and_loan = data.groupby(['Education','Loan_Status']).size().unstack()
education_and_loan
```

```python
#Plotting a stacked bar plot to compare education status and Loan status
education_and_loan.plot(kind= 'bar', stacked = False)

#Changing the x-axis label
plt.xlabel('Education Status')

#Changing the y-axis label
plt.ylabel('Loan Status')

#Rotating the ticks of X-axis
plt.xticks(rotation = 45)
```

```python
#Subsetting the dataframe based on 'Education' column to filter graduates
graduate = data[data['Education'] == 'Graduate']
graduate.head(10)
```

```python
#Subsetting the dataframe based on 'Education' column to filter non graduates
not_graduate = data[data['Education'] == 'Not Graduate']
not_graduate.head(10)
```

```python
#Plotting density plot for 'Graduate' and 'non graduates'
graduate['LoanAmount'].plot(kind = 'density', label = 'Graduate')
not_graduate['LoanAmount'].plot(kind = 'density', label = 'Not Graduate')

#For automatic legend display
plt.legend()
```

```python
#Setting up the subplots
fig , (ax_1, ax_2, ax_3) = plt.subplots(nrows = 3, ncols = 1, figsize = (10,12))

#Plotting scatter plot for Applicant Income and Loan Amount
ax_1.scatter(data['ApplicantIncome'],data['LoanAmount'])

#Setting the subplot 1 axis title
ax_1.set(title = 'Applicant Income')

#Plotting scatter plot for Coapplicant Income and Loan Amount
ax_2.scatter(data['CoapplicantIncome'],data['LoanAmount'])

#Setting the subplot 2 axis title
ax_2.set(title = 'Coapplicant Income')

#Plotting a new column 'TotalIncome'
data['TotalIncome'] = data['ApplicantIncome'] + data['CoapplicantIncome']

#Plotting scatter plot for Total Income and Loan Amount
ax_3.scatter(data['TotalIncome'],data['LoanAmount'])

#Setting the subplot axis title
ax_3.set(title = 'Total Income')
```
